```python
import cv2
import matplotlib.pyplot as plt

# Load the image in grayscale
image = cv2.imread('data/coins.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
_, binary_thresh = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)

adaptive_thresh = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                        cv2.THRESH_BINARY, blockSize=11, C=2)

_, otsu_thresh = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
```

```python
plt.figure(figsize=(10, 8))

# Display original image
plt.subplot(2, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Display binary thresholded image
plt.subplot(2, 2, 2)
plt.imshow(binary_thresh, cmap='gray')
plt.title('Binary Thresholding')
plt.axis('off')

# Display adaptive thresholded image
plt.subplot(2, 2, 3)
plt.imshow(adaptive_thresh, cmap='gray')
plt.title('Adaptive Thresholding')
plt.axis('off')

# Display Otsuâ€™s thresholded image
plt.subplot(2, 2, 4)
plt.imshow(otsu_thresh, cmap='gray')
plt.title('Otsuâ€™s Thresholding')
plt.axis('off')

plt.tight_layout()
plt.show()
```

* Binary Thresholding is fast and straightforward but may not work well for images with varying illumination.
* Adaptive Thresholding adjusts to local variations, making it better for non-uniform lighting but may introduce noise in homogeneous areas.
* Otsuâ€™s Thresholding finds an optimal threshold for global contrast, providing clearer boundaries when the image has distinct object-background contrast.
* On this image because of illumination differences binary thresholding is failing, adaptive thresholding is introducing noise and as there is no distinct object- background contrast due to illumination effect otsu's is also not perfectly thresholding the coins.But it is far better than the other 2.
